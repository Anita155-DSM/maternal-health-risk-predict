```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_prueba = pd.read_csv("maternal-health-risk.csv")

df_prueba.head()
```

```python
df_prueba = df_prueba.drop_duplicates().reset_index(drop=True)

df_prueba.duplicated().sum() #eliminamos duplicados
```

```python
df_prueba.shape
```

```python
df_prueba.hist(edgecolor="black", figsize=(10,10))
plt.suptitle("distribucion de variables")
```

```python
sns.countplot(data=df_prueba, x="RiskLevel", order=["low risk","mid risk","high risk"])
plt.title("conteo por niveles de riesgo")
plt.xlabel("tipos de niveles")
plt.ylabel("cantidad de casos por tipo de nivel")
```
